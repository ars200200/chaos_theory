$$ X_{n+1} = \ln {r(|X_n|)}$$
$$ r \in (0, 3]$$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func(r, x):
    return np.log(r*np.abs(x) +0.001)

# 
n = 1000
r_values = np.linspace(0.01, 3, n)
iterations = 1000
last_iterations = 100

#   x
x = 1e-5 * np.ones(n)

#   
for i in range(iterations):
    x = func(r_values, x)
    if i >= (iterations - last_iterations):
        plt.plot(r_values, x, ',k', alpha=0.25)

plt.xlabel('r')
plt.ylabel('x')
plt.title('Bifurcation diagram')
plt.show()
```

```python
x0 = 5
r = 0.05
plt.grid()
plt.plot(np.arange(-5, 10), np.arange(-5, 10), color='blue')
for k in range(50):
    
    
    plt.plot([x0, x0], [x0, func(r, x0)], color='r')
    plt.plot([x0, func(r, x0)], [func(r, x0), func(r, x0)], color='r')
    x0 = func(r, x0)
```
