```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def f1(x, y, mu):
    return mu * x - y + x * y ** 2

def f2(x, y, mu):
    return x + mu * y + y ** 3

def f(y, t):
    x, y = y
    dydt = [mu * x - y + x * y ** 2, x + mu * y + y ** 3]
    return dydt
```

```python
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)
```

```python
mu = -1
dxdt_negative_mu = f1(X, Y, mu)
dydt_negative_mu = f2(X, Y, mu)


y0 = [0.3, 0.3]

t = np.linspace(0, 10, 1000)

sol = odeint(f, y0, t)

plt.figure()
plt.plot(sol[:, 0], sol[:, 1])
plt.scatter(0, 0, color='blue', label='Точка (0, 0)')
plt.scatter(y0[0], y0[1], color='red', label='Начальная точка (0.3, 0.3)')
plt.quiver(X, Y, dxdt_negative_mu, dydt_negative_mu)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.title('Особая точка устойчивый узел');
```

```python
mu = 1
dxdt_positive_mu = f1(X, Y, mu)
dydt_positive_mu = f2(X, Y, mu)


y0 = [0.3, 0.3]

t = np.linspace(0, 0.7, 1000)

sol = odeint(f, y0, t)

plt.figure()
plt.plot(sol[:, 0], sol[:, 1])
plt.scatter(0, 0, color='blue', label='Точка (0, 0)')
plt.scatter(y0[0], y0[1], color='red', label='Начальная точка (0.3, 0.3)')
plt.quiver(X, Y, dxdt_positive_mu, dydt_positive_mu)

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.title('Особая точка неустойчивый узел');
```

```python
mu = 0
dxdt_zero_mu = f1(X, Y, mu)
dydt_zero_mu = f2(X, Y, mu)


y0 = [0.1, 0.1]

t = np.linspace(0, 10, 1000)

sol = odeint(f, y0, t)

plt.figure()
plt.plot(sol[:, 0], sol[:, 1])
plt.scatter(0, 0, color='blue', label='Точка (0, 0)')
plt.scatter(y0[0], y0[1], color='red', label='Начальная точка (0.3, 0.3)')

plt.quiver(X, Y, dxdt_zero_mu, dydt_zero_mu)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.title('Особая точка эллипс');
```
